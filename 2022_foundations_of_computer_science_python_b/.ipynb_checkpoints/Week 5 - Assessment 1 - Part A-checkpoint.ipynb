```python
# Loading modules
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
import math
from sklearn.datasets import load_diabetes

# Setting figure style and presentation 
sns.set()
%matplotlib inline

#Loading diabetes dataset from scikit-learn module
diabetes = load_diabetes()
diabetes_df = pd.DataFrame(diabetes.data, columns = diabetes.feature_names)
diabetes_df['Disease progression'] = diabetes.target
diabetes_df.duplicated().sum() # none 
print("Note: The data has been centered and scaled,"+
      " after some digging I found the original paper and there is no key to identify \nsex."+
     " This makes the sign of the value (and the column as a whole) completely meaningless,"+
     " without knowing which sex is which we cannot use sex as a predictor of the target variable," +
      " regardless of correlation.")
diabetes_df.isnull().sum()       # No missing values, can't really sense check further with the normalized data
sns.displot(diabetes_df['Disease progression'], bins= 10)
plt.title("Disease progression\n", fontsize=20)
plt.show
```

```python
# Loading modules required for modelling
from sklearn.svm import SVR                             # To access the Support-vector regresssor
from sklearn.model_selection import train_test_split    # To create training/testing sets
from sklearn.model_selection import GridSearchCV        # To creat a tuning grid for hyperparameters

# Loading modules to assess error
from sklearn.metrics import mean_squared_error          # To measure residual variance from line of best fit

# loading modules to assess model
from sklearn.model_selection import cross_val_score     # To determine the model's cross validation accuracy scores
from sklearn.metrics import accuracy_score              # To determine the model's accuracy
from sklearn.dummy import DummyRegressor                # To create a regressor baseline

# Loading modules to plot confusion matrix
import matplotlib.pyplot as plt                         # This one
import seaborn as sns                                   # ... and this one are used to create figure objects
from sklearn.metrics import confusion_matrix            # To create confusion matrix
%matplotlib inline

# Selecting best features 
    # Observing correlation with ehat map
correlation_matrix = diabetes_df.corr()                 # From pandas module
plt.figure(figsize = (12,10))
sns.heatmap(data = correlation_matrix, annot = True, fmt='.2f', linewidths=.5, cmap ='viridis_r')
plt.title("\nCorrelation Matrix presented as a heat map\n", fontsize = 20)
plt.show()
print("Age, sex, s1, and s2 appear to be of little consequence (weakly related)" +
      " to Disease progression and will be exluded\n")

X = pd.DataFrame(np.c_[diabetes_df.bmi, diabetes_df.s3, diabetes_df.s4,
                      diabetes_df.s5, diabetes_df.s6], 
                 columns=['bmi', 's3', 's4','s5','s6'])
y = diabetes_df["Disease progression"]

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42)
print('Training dataset: X_train=', X_train.shape, ', y_train', y_train.shape)
print('Testing dataset: X_test=', X_test.shape, ', y_test', y_test.shape,"\n")

# Creating the model
diabetes_rbf = SVR(kernel = 'rbf', C = 100, gamma = 0.1, epsilon = .1)            # rbf kernel  
print("Before optimization with tuning: ", diabetes_rbf)

# Hyperparameters to calibrate 
param_grid = {'C': [1, 5, 10, 50, 10, 200, 500, 600],
              'gamma': [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 0.9],
              'epsilon': [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.05]}

grid_rbf = GridSearchCV(diabetes_rbf, param_grid, cv = 5)

# Training the model on training data
diabetes_rbf.fit(X_train, y_train)
grid_rbf.fit(X_train, y_train)
print("After optimization with tuning: ", grid_rbf.best_estimator_)

# Assessing model on training and testing sets
   # Creating naive baseline to compare with
#baseline = np.mean(y)
#y_baseline = np.repeat(baseline, len(y_test))

# !!!!! Altered due to comments made in the tute and discussion board, 
# however as its just a point of comparrsion I would argue either approach is valid !!!!

baseline = np.mean(y_train)
y_baseline = np.repeat(baseline, len(y_test))
naive_RSME = mean_squared_error(y_test, y_baseline)
naive_RSME = np.sqrt(naive_RSME)

    # Creating a DummyRegressor baseline
dummy_mean = DummyRegressor(strategy = 'mean').fit(X_train, y_train)
y_predict_dummy_mean = dummy_mean.predict(X_test)
dummy_RSME = mean_squared_error(y_test, y_predict_dummy_mean)
dummy_RSME = np.sqrt(dummy_RSME)

    # Fitting the untuned and tuned model on the testing data
y_test_untuned_rbf = diabetes_rbf.predict(X_test)
y_test_prediction_rbf = grid_rbf.predict(X_test)

    # Measuring RMSE for prediction sets
RMSE_training_rbf = np.sqrt(mean_squared_error(y_test, y_test_untuned_rbf))
RMSE_test_rbf = np.sqrt(mean_squared_error(y_test, y_test_prediction_rbf))
print('\nThe naive RMSE baseline is ', naive_RSME)
print('The dummy regressor RMSE baseline is ', dummy_RSME)
print('The rbf kernel model performance (RMSE)', RMSE_training_rbf)
print('The tuned rbf kernel model performance (RMSE) ', RMSE_test_rbf)
```

```python
# Selecting best features as above - as above
# Splitting the data - as above

# Creating the model
diabetes_linear = SVR(kernel = 'linear', C = 100, gamma = 0.1, epsilon = .1)            # linear kernel doesn't use gamma 
print("Before optimization with tuning: ", diabetes_linear)

# Hyperparameters to calibrate - param_grid from above
grid_linear = GridSearchCV(diabetes_linear, param_grid, cv = 5)

# Training the model on training data
diabetes_linear.fit(X_train, y_train)
grid_linear.fit(X_train, y_train)
print("After optimization with tuning: ", grid_linear.best_estimator_)

# Assessing model on training and test sets
    # Creating baseline to compare with
    # As above

    # Creating a DummyRegressor baseline
dummy_mean = DummyRegressor(strategy = 'mean').fit(X_train, y_train)
y_predict_dummy_mean = dummy_mean.predict(X_test)
dummy_RSME = mean_squared_error(y_test, y_predict_dummy_mean)
dummy_RSME = np.sqrt(dummy_RSME)
    # Fitting the test set using untuned and tuned models
y_test_untuned_linear = diabetes_linear.predict(X_test)
y_test_prediction_linear = grid_linear.predict(X_test)
    # Measuring RMSE of the untuned and tuned models
RMSE_training_linear = np.sqrt(mean_squared_error(y_test, y_test_untuned_linear))
RMSE_test_linear = np.sqrt(mean_squared_error(y_test, y_test_prediction_linear))
print('\nThe naive RMSE baseline is ', naive_RSME)
print('The dummy regressor RMSE baseline is ', dummy_RSME)
print('The linear kernel model performance (RMSE)', RMSE_training_linear)
print('The tuned linear kernel model performance(RMSE) ', RMSE_test_linear)
```

```python
# Plotting the changes in RMSE 
fig = plt.figure(figsize = (20,20))
fig.subplots_adjust( left = 0, right = 1, bottom = 0, top = 1, hspace =0.5, wspace = 0.5)

ax = fig.add_subplot(4,2, 1)
plt.scatter(y_test, y_baseline, color = 'red')
plt.plot(range(0,320), range(0,320),  color = 'green')
plt.title("The naive baseline", fontsize=18)
plt.xlabel('True values', fontsize = 16)
plt.ylabel('Fitted values', fontsize = 16)

ax = fig.add_subplot(4,2, 2)
plt.scatter(y_test, y_baseline, color = 'red')
plt.plot(range(0,320), range(0,320),  color = 'green')
plt.title("The naive baseline", fontsize=18)
plt.xlabel('True values', fontsize = 16)
plt.ylabel('Fitted values', fontsize = 16)

ax = fig.add_subplot(4,2, 3)
plt.scatter(y_test, y_predict_dummy_mean, color = 'magenta')
plt.plot(range(0,320), range(0,320),  color = 'green')
plt.title("The dummy regressor baseline", fontsize=18)
plt.xlabel('True values', fontsize = 16)
plt.ylabel('Fitted values', fontsize = 16)


ax = fig.add_subplot(4,2, 4)
plt.scatter(y_test, y_predict_dummy_mean, color = 'magenta')
plt.plot(range(0,320), range(0,320),  color = 'green')
plt.title("The dummy regressor baseline", fontsize=18)
plt.xlabel('True values', fontsize = 16)
plt.ylabel('Fitted values', fontsize = 16)


ax = fig.add_subplot(4,2, 5)
plt.scatter(y_test, y_test_untuned_rbf, color = 'red')
plt.plot(range(0,320), range(0,320),  color = 'green')
plt.title("The 'rbf' kernel model", fontsize=18)
plt.xlabel('True values', fontsize = 16)
plt.ylabel('Fitted values', fontsize = 16)

ax = fig.add_subplot(4,2, 6)
plt.scatter(y_test, y_test_untuned_linear, color = 'blue')
plt.plot(range(0,320), range(0,320),  color = 'green')
plt.title("The 'linear' kernel model", fontsize=18)
plt.xlabel('True values', fontsize = 16)
plt.ylabel('Fitted values', fontsize = 16)

ax = fig.add_subplot(4,2, 7)
plt.scatter(y_test, y_test_prediction_rbf, color = 'red')
plt.plot(range(0,320), range(0,320),  color = 'green')
plt.title("The tuned 'rbf' kernel model", fontsize=18)
plt.xlabel('True values', fontsize = 16)
plt.ylabel('Fitted values', fontsize = 16)

ax = fig.add_subplot(4,2, 8)
plt.scatter(y_test, y_test_prediction_linear, color = 'blue')
plt.plot(range(0,320), range(0,320),  color = 'green')
plt.title("The tuned 'linear' kernel model", fontsize=18)
plt.xlabel('True values', fontsize = 16)
plt.ylabel('Fitted values', fontsize = 16)
print("\nThe RMSE is a commentary on how close each point is to the green line." +
      " Ideally the fitted values are equivalent to the true values\n")
```

### Summary
**Problem defenition**

The problem was outlined in the assessment. We are to create a support-vector regression model that can predict disease progression based on a set of predictor features.

**Data ingestion**

The diabetes dataset was imported from the scikit-learn module along with it's outcome values.

**Data Cleaning**

No missing values were observed and cleaning was made difficult as the dataset is already normalized so we can't really do a sense check (eg, heart beat > 0, etc). The outcome values were not normalized, but my research indicates that was not a concern for this model. Additionally, the target variable had a significant right skew, but my research also determined that log transformations were not necessary for support-vector machines.

A corrolation matrix was used to identify the degree of correlation between predictors and the target variable. I decided on a threshold of more than 0.25 or less than -0.25.

**Data splititng**

The data was split into testing and training sets. The training set was also broken into validation sets during the tuning phase (GridSearchCV).

**Model Training**

The support vector machine was modelled with the rbf kernel and then again seperately with the linear kernel on the same training set. These models were then tuned to optimize their parameters to better fit the data. 

**Model Selection**

A naive baseline was produced using the mean value of the training sets outcome variable as discussed in the tute. A dummy regression baseline was created using the median value of predicted values using the test data. The untuned fitted model for each kernel can be used as a baseline to measure the effects of tuning parameters. The naive baseline, dummy regressor baseline, untuned models, and tuned models were used to create fitted predictions. These predictions were used in conjunction with the true values to calculate the RMSE values. The error has also been plotted with the green line indicating a model with 100% accuracy.

The two baselines had approximately the same RMSE values, likely because the dummy regressor was set to mean as well. The two tuned models had very similar RMSE values; the 'rbf' kernal producing only a slightly lower RSME value than the 'linear' kernel. The 'rbf' kernel had the best improvement after tuning, however the untuned 'linear' model performed far better than the 'rbf's untuned model. When comparing the tuned 'rbf' model to the baselines, there is clearly an improvement. The same can be said of the tuned 'linear' model. However, I would not recommend either of these models be used to predict the progression of disease. The error is far too high to make predictions regarding a persons health.

```python
from sklearn.ensemble import RandomForestRegressor #random forrest
# Creating model
rf_model = RandomForestRegressor(n_estimators = 2000, random_state = 42)
print("Before optimization with tuning: ", rf_model)
```

```python
X = pd.DataFrame(diabetes.data, columns = diabetes.feature_names)
y = diabetes_df["Disease progression"]

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42)
print('Training dataset: X_train', X_train.shape, ', y_train', y_train.shape)
print('Testing dataset: X_test', X_test.shape, ', y_test', y_test.shape,"\n")

# Tuning hyper params
grid_rf = {'max_depth': [int(x) for x in np.linspace(1, 10, num = 2)],
           'max_features': [5, 6, 7, 8]}
untuned_rf = rf_model.fit(X_train, y_train)
tuned_rf = GridSearchCV(rf_model, grid_rf, cv = 5)


# fitting model
untuned_fit = untuned_rf.fit(X_train, y_train)
tuned_fit = tuned_rf.fit(X_train, y_train)
print("After optimization with tuning: ", tuned_fit.best_params_, "\n")

# fitting predicting values of test set
prediction_rf_untuned = untuned_fit.predict(X_test) 
prediction_rf_tuned = tuned_fit.predict(X_test) 

# Plotting model
fig = plt.figure(figsize = (20,5))
fig.subplots_adjust( left = 0, right = 1, bottom = 0, top = 1, hspace =0.5, wspace = 0.5)

ax = fig.add_subplot(1,2, 1)
plt.scatter(y_test, prediction_rf_untuned
            , color = 'blue')
plt.plot(range(0,320), range(0,320),  color = 'green')
plt.title("The untuned random forest regression model", fontsize=18)
plt.xlabel('True values', fontsize = 16)
plt.ylabel('Fitted values', fontsize = 16)

ax = fig.add_subplot(1,2, 2)
plt.scatter(y_test, prediction_rf_tuned
            , color = 'blue')
plt.plot(range(0,320), range(0,320),  color = 'green')
plt.title("The tuned random forest regression model", fontsize=18)
plt.xlabel('True values', fontsize = 16)
plt.ylabel('Fitted values', fontsize = 16)

# Checking RMSE
untuned_mse = mean_squared_error(y_test, prediction_rf_untuned)
untuned_rmse = np.sqrt(untuned_mse)
tuned_mse = mean_squared_error(y_test, prediction_rf_tuned)
tuned_rmse =np.sqrt(tuned_mse)
print("The untuned random forest model had a RMSE of ", untuned_rmse,
      " and the tuned random forest model had a RMSE of ", tuned_rmse,
      "not a great improvement on the Support vector machine, but a great deal more time consuming")
```
