```python
# Loading modules required for data manipulation an EDA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
%matplotlib inline
```

```python
# Importing Data
hayden_df = pd.read_csv('all_features.csv')
hayden_df.head()
```

```python
# Creating a plotly figure in order to easily identigy points of interest
fig = go.Figure()
fig.add_trace(go.Scatter(x = hayden_df['publish date'], y = hayden_df['times liked'],
                         mode='markers', name = 'Likes'))
#fig.add_trace(go.Scatter(x = hayden_df['publish date'], y = hayden_df,
#                         mode='markers', name = 'Views'))
# plt.scatter(yt_vids['publish date'], yt_vids['times favourited'] ), only 0s
fig.update_layout(title_text = "Likes for each YouTube video",
                  yaxis_title = "Count",
                  xaxis_title = "Time",
                       showlegend=True)
```

```python
hayden_df.columns
```

```python
# Creating a bar chart to demonstarte upload regularity coloured by likes recieved for each video
months_uploaded = px.bar(hayden_df, x = "month uploaded",  
                         color = "times liked", 
                         title = "Videos uploaded per month coloured by the number of lieks recieved")

    # Ordering x-axis chronologically
months_uploaded.update_xaxes(categoryorder='array', categoryarray= ['January', 'February', 'March', 'April', 'May',
                                                                    'June', 'July', 'August', 'September', 'October',
                                                                    'November', 'December'])
months_uploaded.show()
```

```python
# Recreating 'day uplaoded' column
days = pd.melt(hayden_df, id_vars = ['video ids', 'times liked'], 
        value_vars=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], 
        var_name='day uploaded', 
        value_name='value')
valid_days = days.loc[days['value'] == 1.0]
valid_days

# Creating a bar chart to demonstrate weekly upload habits
day_uploaded = px.bar(valid_days, 
                      y = ['day uploaded'],  
                      color = "times liked", 
                      title = "Upload habits with instances coloured by total likes",
                      labels = {'count': 'Videos uploaded', "_value" : "Day of upload", "times liked" : "Times liked"})
day_uploaded.show()
```
