### Section 1 -  Inheritance and Polymorphism
Below I have created three classes; a team super class and coach and player subclasses. I have set them up so the have a name function which is polymorphic in nature as it can be called by each type of object but print a different response in each case. I have also provided the code needed for a sorting function to work on a collection of Player objects.

```python
import random
# Creating classes with inheritance and polymorphism, players will be sorted by age. Coaches could be sort by experience
class Team():
    """ A Team class thatc reates objects with attributes of team name, number of players, and the coach's name."""
    def __init__(self, name, coach):
        self.name = name
        self.coach = coach
        
    def named(self):
        print("The team name is "+ self.name.title())
    
class Player(Team):
    """ A Player class that inherits from the Team class and has name, age, position, team attributes."""
    def __init__(self, name, age, position, coach, team):
        self.name = name
        self.age = age
        self.position = position
        self.coach = coach
        self.team = team
        
    def named(self):
        print ("The player's name is " + self.name.title())
    # function to return the string of the object
    def __repr__(self):
        return str(self.age)
    # Less than function
    def __lt__(self, other):
        return self.age < other.age
    # greater than function
    def __gt__(self, other):
        return self.age > other.age
    # Equal to function
    def __eq__(self,other):
        return self.age == other.age
    
class Coach(Team):
    """ A Coach class that inherits from the Team class and has name and experience attributes."""
    def __init__(self, name, experience):
        self.name = name
        self.experience = experience
    def named(self):
        print ("the coach's name is "+ self.name.title())

# Creating an empty list for the collection
player_list=[]
# Adding 10 players to the list
for i in range(10,0,-1):
    # Creating player attributes
    name = "Player" + str(i)
    age = i*2 +15
    position = random.choice(["Forward", "Midfield", "Backline", "Goalkeeper"])
    coach = "Payal"
    team = "Adelaide university Soccer Team"
    # Creating an instance of a Player 
    player = Player(name, age, position, coach, team)
    # Adding the player to the list
    player_list.append(player)

print(player_list)
```

```python
# Demonstrating the sort function is functional
player_list.sort()
print("List in ascending order: ", player_list)
player_list.sort(reverse = True)
print("List in descending order: ", player_list)
```

```python
# Demonstrating polymorphism in the named() function on three different object types
print("\nDemonstrating polymorphism")
team1 = Team("Adelaide University", "Payal")
player1 = Player("Dylan", 25, "Midfield", "Payal", "Adelaide University")
coach1 = Coach("Payal", "10 years")

team1.named()
player1.named()
coach1.named()
```

### Section 2 -  Union (concat and append)
The union of two dataframes is effectively just adding rows from one dataframe to another dataframe when they have matching column names. This can be done with the concat and append functions. I have demonstrated and  timed each of these below:

```python
import pandas as pd
import numpy as np
# Loading two dataframes for complexity marks
user_devices = pd.read_csv("https://github.com/shanealynn/Pandas-Merge-Tutorial/blob/master/user_device.csv?raw=true")
print("Information regarding user devices. Dataframe takes the shape ", user_devices.shape)
user_devices.head()
```

```python
user_usage = pd.read_csv("https://github.com/shanealynn/Pandas-Merge-Tutorial/blob/master/user_usage.csv?raw=true")
print("Information regarding usage of device. Dataframe has the shape ", user_usage.shape)
user_usage.head()
```

```python
# Union Method 1 - concat()
pd.concat([user_devices, user_usage], join = 'outer', sort = False)
```

```python
# Checking average duration
%timeit pd.concat([user_devices, user_usage], join = 'outer', sort = False)
```

```python
# Union Method 2 - append()
df = pd.DataFrame()
df.append([user_devices, user_usage], sort = None)
```

```python
# Checking average duration
df = pd.DataFrame()
%timeit df.append([user_devices, user_usage], sort = None)
```

### Section 3 -  Merge (Merge and join)
This is effectively combining instances that are spread across two dataframes into one. This can be accomplished with merge and join functions. I have demonstrated and time each below.

```python
# Merge Method 1 - merge()
pd.merge(user_devices, user_usage)
```

```python
# Checking average duration
%timeit pd.merge(user_devices, user_usage)
```

```python
# Merge Method 2 - join()
user_devices.set_index('use_id').join(user_usage.set_index('use_id'), how = 'inner', lsuffix ="_l", rsuffix = '_r')

# Had to assign a column to be the index in each dataframe for this to work smoothly 
```

```python
# Checking average duration
%timeit user_devices.set_index('use_id').join(user_usage.set_index('use_id'), how = 'inner', lsuffix ="_l", rsuffix = '_r')
```

### Summary
In comparing the average processing time, it appeared that the merge functions took longer than the union functions. Merge() took longer than join(); and append() took longer than concat(), but not by much.
