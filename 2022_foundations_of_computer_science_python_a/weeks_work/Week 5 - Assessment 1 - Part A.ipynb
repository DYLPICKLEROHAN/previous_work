### Q1 -  Plotting Panda Data Frames with Date indexes 
Using the numpy modulke, below I introduce 6 diffferent series that are indexed by datetime values from 2016-01-01 until now with values randomly selected between 0 and 1. I then combine these series as columns in a DataFrame using the pandas module. The number of days between 1/1/2016 and now have been printed along with the first 30 rows of the DataFrame. I then proceed to create a dataframe of the cumulative sum of values in each column and print the first five rows as directed. The values have been presented in a line chart, and individually in subplots (using two different methods). Then in order to quickly plot the accumulative sum of each column, I used an area chart which combines all the values at each indexed point in time.

These charts reveal an obvious positive trend over time. As the y value increases by sum positive value from 0 to (but not including) 1, it's unsurpring. They all appear similar over large periods of time because the larger the domain, the more the trend will follow the average. In terms of the area chart, the gradient is mathematically related to it constituents and so it will be roughly six times the gradient of any individual series (as it took the accumulation of the six of themto generate it). I could have combined the collumns mathematically into a new 1D series, but this generates the same results mathematically and gives more information.  

```python
# Question 1.1)
#Create six time series stored in a Pandas data frame:
#a) The data frame must have an index that is a range of dates from 2016-01-01 until today.
#b) Each column contains one set of random numbers in a range 0 to 1 (each column will have
#one random number for each date).

# Importing the numpy and pandas modules
import numpy as np
import pandas as pd

# Generating an index
date_time_index = pd.date_range('2016-01-01', 'now') 
series_length = len(date_time_index)

# Creating six time series containing random values between [0,1) and indexing them with the timestamp:
timeseries_1 = pd.Series(np.random.rand(series_length), index = date_time_index)
timeseries_2 = pd.Series(np.random.rand(series_length), index = date_time_index)
timeseries_3 = pd.Series(np.random.rand(series_length), index = date_time_index)
timeseries_4 = pd.Series(np.random.rand(series_length), index = date_time_index)
timeseries_5 = pd.Series(np.random.rand(series_length), index = date_time_index)
timeseries_6 = pd.Series(np.random.rand(series_length), index = date_time_index)

# Compiling the 6 series into a DataFrame
time_frame = pd.DataFrame({"series 1":timeseries_1, "series 2":timeseries_2, "series 3":timeseries_3,
                           "series 4":timeseries_4, "series 5":timeseries_5, "series 6":timeseries_6})

###############################################################################################################
# Question 1.2) Calculate and print the number of days from 2016-01-01 until today.
print("The number of days since 01/01/2016 is: ", series_length)

#################################################################################################################
# Question 1.3) Show the first 30 rows of your data 
time_frame.head(30)
```

```python
# Question 1.4) Calculate the cumulative sum of the Data frame
# Calculating the cumulative sums and presenting the first five rows
df_accumulate = time_frame.cumsum()
df_accumulate.head(5)
```

```python
# Question 1.5) Plot the six sets of row values of df_accumulate in a graph
# Ensuring all modules are loaded
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline 
# magic function enabling plots to appear on notebook

# Plotting and styling the six different series
figure1 = plt.figure()
df_accumulate.plot(style =['r-','g--','b:','m-.','k:','c:'], grid = True)
plt.title("Figure 1: The six sets of accumulated values plotted over time", fontsize =15, pad = 20)
plt.xlabel("Time (Years)")
plt.ylabel("Accumulative value (Units)")
plt.text(17500,100, "It makes sense, at most they could\n" 
         "only increase by <1 unit in y per day", fontsize = 10)
```

```python
# Question 1.6)
# Plot the six sets of row values of df_accumulate in separate graphs using subplots 
# then plot the accumulate sum of each of the columns (random numbers).

# Ensuring all modules are loaded
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# Method 1 - plotting simple subplots
figure2 = plt.figure()
axs = df_accumulate.plot(subplots=True, 
                         style= ['r-','g--','b:','m-.','k:','c:'],
                         fontsize=10, 
                         layout =(2,3), 
                         figsize = (20,10), 
                         grid =True)

plt.setp(axs[:, :], xlabel='Time (years)')
plt.setp(axs[:, :], ylabel='Accumulated value (units)')
plt.suptitle("Figure 2: Accumulated Series values for series 1 to 6", fontsize=20)
axs[0, 0].set_title("Series 1", fontsize = 12, pad = 20)
axs[0, 1].set_title("Series 2", fontsize = 12, pad = 20)
axs[0, 2].set_title("Series 3", fontsize = 12, pad = 20)
axs[1, 0].set_title("Series 4", fontsize = 12, pad = 20)
axs[1, 1].set_title("Series 5", fontsize = 12, pad = 20)
axs[1, 2].set_title("Series 6", fontsize = 12, pad = 20)
```

```python
# Method 2 - plotting more complex subplots
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

number = np.arange(0., 5., 0.2)
legend = ['series 1','series 2','series 3', 'series 4', 'series 5', 'series 6']
figure3 = plt.figure()
plots = df_accumulate.plot(style = ['r-','g--','b:','m-.','k:','c:'], 
                           fontsize = 10,
                           figsize=(10,5))
plots.legend(legend, loc='upper left')
plt.grid(True)
plt.xlabel("Time (Years)", fontsize =15)
plt.ylabel("Accumulated value (units)", fontsize =15)
plt.title("Figure 3: The accumulated values for series 1 to 6", pad =20, fontsize=20)

figure4 = plt.figure()
rows = 3
columns =2
style =[0, 'r-','g--','b:','m-.','k:','c:']
fig, plots = plt.subplots(rows, columns)

counter=1 
for r in range(rows):
    for c in range(columns):
        df_accumulate["series "+str(counter)].plot(ax=plots[r,c], 
                                                   figsize=(10,10), 
                                                   title = "Series "+str(counter), 
                                                   style= style[0+counter], 
                                                   xlabel="Time (Years)", 
                                                   ylabel= "Accumulated Value (Units)",
                                                   grid =True)
        counter= counter + 1
plt.suptitle("Figure 4: Subplot of accumulated values for series 1 to 6", fontsize= 20)
fig.tight_layout(pad=3)
```

### Question 4 - Cleaning Data
Below I have cleaned the responses to the question of Gender into three categories. I then create dataframes for males and females that indicate the number or people that have sought treatment for mental illness in both cases. These data sets are then used to create pie charts.

A larger portion of the female population were found to have sought treatment than that of the male population. The population as a whole was found to have 50% of subjects reporting they had sought treatment. So although males appear less likely to seek treatment, the entire population has an issue seeking support. In general, it seemed those who did not identify as Cis-gendered were most likley to seek treatment. However, the population of subjects was also heavily male dominant (988 of 1254). It is difficult to say how reliable these results are when you compare a population of 247 women to one of 988 men, not to mention the 19 subjects recorded as other.

```python
# Question 4.a)
# Unifying the responses to Gender into three categories: Male, Female, Other
identified_male = ['male','m', 'M', 'Make', 'Man', "Mail", "Malr",'msle','Mal',
                   'maile', "Male", 'Cis Male', 'Male (CIS)','cis male', 'Cis man', "Male ",'Cis Man','Guy (-ish) ^_^']
identified_female = ['female','F','f','Female','Femake','woman','femail','Woman', 
                     'Cis Female', 'cis-female/femme', "Female (cis)", "Female "]
identified_other = ['Male-ish', 'Trans-female', 'something kinda male?', 'queer/she/they',
                    'non-binary', 'Nah', 'Enby', 'fluid', 'Genderqueer', 'Androgyne',
                    'Agender', 'male leaning androgynous', 'Trans woman', 'Neuter',
                    'Female (trans)', 'queer', 'A little about you', 'ostensibly male, unsure what that really means']

for value in identified_male:
    data.loc[(data.Gender == value,'Gender')]='Male'
    
for value in identified_female:
    data.loc[(data.Gender == value,'Gender')]='Female'

for value in identified_other:
    data.loc[(data.Gender == value,'Gender')]='other'

# I did notice another two options not listed: "Male " and "Female "
print(data.Gender.unique())
print(data.treatment.unique())
print("number of people that sought treatment: ", len(data[data["treatment"]=="Yes"]))
print("number of people that responded with 'Yes' or 'No': ", len(data[data["treatment"]=="Yes"])+len(data[data["treatment"]=="No"]))
print("Proportion of affirmative responses: ", round(633/1254*100))
print("number of females", len(data[data["Gender"]=="Female"]))
print("number of males", len(data[data["Gender"]=="Male"]))
print("number of other", len(data[data["Gender"]=="other"]))
```
