### Underlying mechanisms
Computers cannot store an infinte number of decimal places, this leads to **rounding off errors**.
computer components are memory, CPU, and devices. CPU carries out computer instructions. 
CPU takes the operand from the memory, evaluates the result with that operand, and then stores the result back in the memory. Each memory unit is capable of storing a finite amount of information. There is an indexed location thats refferd to a the **memory address**. The devices are the things that connect it to reality, keyboards, mouse, monitor, hard disk, etc.

Computers recognise signals as discrete numbers/integers that get processed and returned to us again as physical signals. Devices act as channels between computers and the physical world, a way for discrete variables to be returned as physical signals (sounds, colours, shapes, etc).
Storage units are typically binary. 0 or 1, n basic storage units can lead to 2^n states. 
00, 01, 11, and 10. These can be converted to binary numbers. The order of 0s and 1s has a value so even lots of information can be stored in something like the branches of a probability tree. 
each binary digit (0 or 1) is called a bit, 8 bits is a byte, so a byte can store 2^8 = 256 distinct binary values. Most computers take 8 bytes/ 64 bits as a **word** 
2's complement interpretation is a method of representing negative numbers. You invert the binary string so 1s become 0 and vbise versa and then add 1 to the final digit. A final digit of 1 indicates a negative number. 


-13 can be written as 00001101 becomes 11110010, then add 1.
11110011. This way negative numbers can still be stored in 8 bits without needing to add an additional bit representing the positive and negative sign. it saves space.
This way when you add 13 to -13, the sum becomes a 9 bit string which will be intrepreted as 0. The correct answer.


**Floating point numbers** are useful when dealing with sound waves. You split the total number of bits in two parts, one fore the *base number*, and one for the *significant*. It takes the form of scientific notation (baseEsignificant).

Because computers operate on 64bit strings, they cannot output more than that, leading to rounding off errors. Numbers can also be intrepretted as long type which allows large integer values, but not floats. Python settles for 15 decimal places to balance computation time and accuracy.

### Vectorization
Uses optimized, pre-compiled code written in a low-level coding language to perform mathematical operations iteratively and more quickly than the native languages coding method. They are done with the unfunc operations.
1. Addition
1. Subscription
1. Negative
1. Multiplication
1. Division
1. Integer division
1. Exponentiation
1. Moduloc (division remainder

There are many more, some of which are nan safe (empty values are ignored).

In any case it is far quicker to use vectorization than itterate through the usual python language.

```python
#Addition
a1 = np.arange(9).reshape(3,3)
a2 = np.arange(3)
print( a1)
print()
print( a2)
print()
print('Addition a1 + 5 =\n', a1+5)
print ('np.add a1 + 5 =\n', np.add(a1,5))
#Addition two arrays 
print ('np.add a1 + a2 =\n', np.add(a1,a2))
```

### Time Series data
Pandas uses timestamped data to create time series data. The date and time can act as the row indexing.
1. timestamps - moment in time
1. timedeltas - a duration of time with a specific start and end, they act as an offset.
1. periods - a reoccuring timedeltas, eery day, every minute, etc.

You us a datetimeindex to merge them with a DataFrame.

```python
import numpy as np
import pandas as pd
import datetime

# creating timestamps
print(pd.Timestamp(2022, 1, 24, 8, 12))

print(pd.Timestamp("now"))

print(pd.Timestamp('24 january 2022 6:13 pm'))
```

```python
# creatng a timedelta
print(pd.Timedelta(weeks =1, days = 3, hours = 4, seconds= 1, milliseconds = 6, microseconds=5))

print(pd.Timedelta('1 day  2 hours'))

delta = pd.Timedelta(1)
time = pd.Timestamp('now')
print(time + delta)
```

```python
# creating a Period (timespan)
# Y, H, M, Q
period = pd.Period(freq='Q', year = 2022, month = 4, quarter = 2, day = 11, hour=8, minute=30, second=5)
print(period)
print(period+1)
print(period+2)

#viewing frequency in other units
print(period.asfreq('min'))
```

```python
#here we use the spearman correlation coefficient from the scipy statistics library
from scipy.stats import spearmanr 

#Filtering the numeric columns (attributes)
#use a for loop to iterate over all of the columns in the data
#if the item is numeric, store its value in numerical_columns
#(n automatically increases)
numerical_columns = [n for n,col in df.items() if pd.api.types.is_numeric_dtype(col)]
numerical_columns
```

```python
#Getting the correlations between pairs of the numerical columns 
#the dropna() function of Pandas Data Frame drops all of the missing values
corr = spearmanr(df[numerical_columns].dropna()).correlation
corr
```

```python
plt.figure(figsize=(16,12))

#in this case we have set the x and y labels as they are not part of the data
# corr only holds the values in an array (see above)
sns.heatmap(corr, annot=True, fmt='.2f', linewidths=.5,
           xticklabels=numerical_columns, yticklabels=numerical_columns) 
```
